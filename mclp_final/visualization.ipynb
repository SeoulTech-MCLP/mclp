```python
import pandas as pd
import folium
from folium.plugins import MiniMap

# 여러 개의 좌표 데이터 예시
df = pd.read_csv('map_points_data.csv', encoding='cp949')
df2 = pd.read_csv('opts_sites_data.csv', encoding='cp949')
df3 = pd.read_csv('fixed_point_data.csv', encoding='cp949')

# print(df)
# print(df2)    # 요놈만 위도-경도가 뒤바뀌어있음
# print(df3)


# 지도의 중심을 설정할 좌표
map_center = [df['위도'].mean(), df['경도'].mean()]


# Folium 지도 생성
my_map = folium.Map(location=map_center, zoom_start=12)


# 행정동 중복 체크를 위한 집합(set) 생성
visited_districts = set()

# map_points_data.csv 파일의 좌표를 지도에 녹색 점으로 추가
for index, row in df.iterrows():
    district = row['행정동']
    if district not in visited_districts:
        visited_districts.add(district)
        
        # 중심 좌표 찾기 (행정동의 모든 좌표의 평균 사용)
        district_coords = df[df['행정동'] == district][['위도', '경도']]
        center_coords = [district_coords['위도'].mean(), district_coords['경도'].mean()]

        folium.Marker(location=center_coords,
                      icon=folium.Icon(color='red'),
                      tooltip=district).add_to(my_map)
        
#     folium.CircleMarker(location=(row['위도'], row['경도']),
#                         radius=1,
#                         color='green',
#                         fill=True,
#                         fill_color='green').add_to(my_map)

# opts_sites_data.csv 파일의 좌표를 지도에 빨간색 점으로 추가
for index, row in df2.iterrows():
    folium.CircleMarker(location=(row['경도'], row['위도']),    # 요놈만 위도-경도 뒤바뀌어있어서 이렇게
                        radius=3,
                        color='red',
                        fill=True,
                        fill_color='red').add_to(my_map)

# fixed_point_data.csv 파일의 좌표를 지도에 파란색 점으로 추가
for index, row in df3.iterrows():
    folium.CircleMarker(location=(row['위도'], row['경도']),
                        radius=1,
                        color='blue',
                        fill=True,
                        fill_color='blue').add_to(my_map)


# 지도 표시
my_map
```

```python
import pandas as pd
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import MiniMap

# 여러 개의 좌표 데이터 예시
df = pd.read_csv('map_points_data.csv', encoding='cp949')
df2 = pd.read_csv('opts_sites_data.csv', encoding='cp949')
df3 = pd.read_csv('fixed_point_data.csv', encoding='cp949')

# print(df)
# print(df2)    # 요놈만 위도-경도가 뒤바뀌어있음
# print(df3)


# 지도의 중심을 설정할 좌표
map_center = [df['위도'].mean(), df['경도'].mean()]


# Folium 지도 생성
my_map = folium.Map(location=map_center, zoom_start=12)

# 행정동 중복 체크를 위한 집합(set) 생성
visited_districts = set()

# map_points_data.csv 파일의 좌표를 지도에 녹색 점으로 추가
for index, row in df.iterrows():
    district = row['행정동']
    if district not in visited_districts:
        visited_districts.add(district)
        
        # 중심 좌표 찾기 (행정동의 모든 좌표의 평균 사용)
        district_coords = df[df['행정동'] == district][['위도', '경도']]
        center_coords = [district_coords['위도'].mean(), district_coords['경도'].mean()]

        # 행정동 경계 표시 (원 모양)
        circle_color = 'orange'
        Circle(location=center_coords,
               radius=1000,  # 반경 조절 가능
               color=circle_color,
               fill=True,
               fill_color=circle_color,
               fill_opacity=0.1).add_to(my_map)
        
#     folium.CircleMarker(location=(row['위도'], row['경도']),
#                         radius=1,
#                         color='green',
#                         fill=True,
#                         fill_color='green').add_to(my_map)

# opts_sites_data.csv 파일의 좌표를 지도에 빨간색 점으로 추가
for index, row in df2.iterrows():
    folium.CircleMarker(location=(row['경도'], row['위도']),    # 요놈만 위도-경도 뒤바뀌어있어서 이렇게
                        radius=3,
                        color='red',
                        fill=True,
                        fill_color='red').add_to(my_map)

# fixed_point_data.csv 파일의 좌표를 지도에 파란색 점으로 추가
for index, row in df3.iterrows():
    folium.CircleMarker(location=(row['위도'], row['경도']),
                        radius=1,
                        color='blue',
                        fill=True,
                        fill_color='blue').add_to(my_map)


# 지도 표시
my_map
```
